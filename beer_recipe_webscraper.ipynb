```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
```

```python
r = requests.get('https://www.brewersfriend.com/homebrew-recipes/')
```

```python
soup = BeautifulSoup(r.content, 'lxml')
```

```python
table = soup.find_all('div', class_='ingredient-page-content')
```

```python
titles = table[0].find_all('span', itemprop='name')
title = [x.text for x in titles]

styles = table[0].find_all('span', itemprop='recipeCuisine')
style = [x.text for x in styles]

sizes = table[0].find_all('span', itemprop='recipeYield')
size = [x.text for x in sizes]

abv = [x.find_next('td').text for x in sizes]
ibu = [x.find_next('td').find_next('td').text for x in sizes]
og = [x.find_next('td').find_next('td').find_next('td').text for x in sizes]
fg = [x.find_next('td').find_next('td').find_next('td').find_next('td').text for x in sizes]
color_soup = [x.find_next('td').find_next('td').find_next('td').find_next('td').find_next('td') for x in sizes]
color = [x.text for x in color_soup]
method = [x.find_next('td').text for x in color_soup]
views = [x.find_next('td').find_next('td').text for x in color_soup]
brewed = [x.find_next('td').find_next('td').find_next('td').text for x in color_soup]


# for second table of each beer:
table = soup.find_all('tr', {'style':"display:none;overflow:hidden;"})

b_title = []
b_author = []
b_bsize = []
b_btime = []
b_bgravity = []
b_efficiency = []
b_mash = []
b_sugar = []
b_brew = []
b_pitch = []
b_ptemp = []
b_pmethod = []
b_pamount = []
b_creation = []
b_notes = []

for beer in table:
    b = [x.text.strip() for x in beer.find_all('td', bgcolor='white')]
    b_title.append(b[0])
    b_author.append(b[1])
    b_bsize.append(b[2])
    b_btime.append(b[3])
    b_bgravity.append(b[4])
    b_efficiency.append(b[5])
    b_mash.append(b[6])
    b_sugar.append(b[7])
    b_brew.append(b[8])
    b_pitch.append(b[9])
    b_ptemp.append(b[10])
    b_pmethod.append(b[11])
    b_pamount.append(b[12])
    b_creation.append(b[13])
    b_notes.append(b[14])
```

```python
# parsing page number 2 and the following
for page in range(2,10):
    r = requests.get('https://www.brewersfriend.com/homebrew-recipes/page/{}/?sort=breweddown'.format(page))
    soup = BeautifulSoup(r.content, 'lxml')
    table = soup.find_all('div', class_='ingredient-page-content')
    
    titles = table[0].find_all('span', itemprop='name')
    title_new = [x.text for x in titles]
    title.extend(title_new)
    
    styles = table[0].find_all('span', itemprop='recipeCuisine')
    style_new = [x.text for x in styles]
    style.extend(style_new)
    
    sizes = table[0].find_all('span', itemprop='recipeYield')
    size_new = [x.text for x in sizes]
    size.extend(size_new)
    
    abv_new = [x.find_next('td').text for x in sizes]
    ibu_new = [x.find_next('td').find_next('td').text for x in sizes]
    og_new = [x.find_next('td').find_next('td').find_next('td').text for x in sizes]
    fg_new = [x.find_next('td').find_next('td').find_next('td').find_next('td').text for x in sizes]
    color_soup = [x.find_next('td').find_next('td').find_next('td').find_next('td').find_next('td') for x in sizes]
    color_new = [x.text for x in color_soup]
    method_new = [x.find_next('td').text for x in color_soup]
    views_new = [x.find_next('td').find_next('td').text for x in color_soup]
    brewed_new = [x.find_next('td').find_next('td').find_next('td').text for x in color_soup]
    abv.extend(abv_new)
    ibu.extend(ibu_new)
    og.extend(og_new)
    fg.extend(fg_new)
    color.extend(color_new)
    method.extend(method_new)
    views.extend(views_new)
    brewed.extend(brewed_new)
    
    
    table = soup.find_all('tr', {'style':"display:none;overflow:hidden;"})
    
    for beer in table:
        b = [x.text.strip() for x in beer.find_all('td', bgcolor='white')]
        b_title.append(b[0])
        b_author.append(b[1])
        b_bsize.append(b[2])
        b_btime.append(b[3])
        b_bgravity.append(b[4])
        b_efficiency.append(b[5])
        b_mash.append(b[6])
        b_sugar.append(b[7])
        b_brew.append(b[8])
        b_pitch.append(b[9])
        b_ptemp.append(b[10])
        b_pmethod.append(b[11])
        b_pamount.append(b[12])
        b_creation.append(b[13])
        b_notes.append(b[14])

    
    # time sleep of 2 seconds to avoid being blocked by the page in case you make too many requests too fast
    time.sleep(2)
```

```python
data = pd.DataFrame()
data['title'] = title
data['style'] = style
data['size'] = size
data['abv'] = abv
data['ibu'] = ibu
data['og'] = og
data['fg'] = fg
data['color'] = color
data['views'] = views
data['brewed'] = brewed
```
